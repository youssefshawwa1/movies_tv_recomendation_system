# Data Wrangling

I saw that all the datasets contain 12 columns and they are consitant. They all have same columns names, such as description, listed_in, type

I realised that I have to combine them into one dataset.

But the problem the show_id is repeated, so I got an idea of prefixing the show_id with its own platform name. This way I get rest of the repetatoin of the show_id.

Also adding a new column I will name it platform, to know what platform this title came from.

```python
# Prefix the show_id with its coresponding platform.

netflix['show_id'] = "nf_" + netflix['show_id']
disney['show_id'] = "dp_" + disney['show_id']
hulu['show_id'] = "hl_" + hulu['show_id']
amazon['show_id'] = "amzn_" + amazon['show_id']

# Adding the platform column

netflix['platform'] = 'Netflix'
disney['platform'] = 'Disney Plus'
hulu['platform'] = 'Hulu'
amazon['platform'] = 'Amazon Prime'



# Making sure that changes are effective

print(netflix[['show_id', 'platform']].head(2))
print(disney[['show_id', 'platform']].head(2))
print(hulu.loc[:1, ['show_id', 'platform']])
print(amazon.loc[:1, ['show_id', 'platform']])
```

Now I need to combine them into one big dataset.
But there is a problem as there might be titles that in more than one platform, so I need first to explore the new dataset and see what I can do.

```python
# Combining all datasets into one dataset.

all_titles = pd.concat([netflix, disney, hulu, amazon])
```

## 6. Visualize the description to make sure its ready for training.

In this part I need to check if there are enough meaning words in the description.

My approach is to first making all the descriptions one giant text, and then visualize the most frequent words.

I will be using a librarry called wordcloud.


Also as I searched, it appears that its better to remove uselses words such as the, is, are...

In order to remove them, I can't type them all in a list, I will be using wordcloud librarry.

```python
from collections import Counter

from utils import clean_description, clean_title, clean_genres

# creating a set of the stopwords (the, is, in...)


# emppty list to store all filtered words from all descriptions
all_words = []
for description in all_titles_clean['description'].astype(str):
    # loop through each descriptoin, and ensure that they are treated asstring.

    # convert description to lowercase, and split the words into a list.
    cleaned_text= clean_description(description)
    # adding the words that are not in the stop_words to the all_words list.
    all_words.extend(cleaned_text.split())


# creating an object of Counter, it counts the accurance of the words such as 'love': 100

word_counts = Counter(all_words)

# getting top 20 words that most common
top_20_words = word_counts.most_common(20)


# - `zip(*...)`: "Unzips" the list of tuples into two separate tuples
# - Example: [("family", 500), ("life", 300)] â†’ ("family", "life") and (500, 300)
# - `words` becomes a tuple of word strings
# - `counts` becomes a tuple of frequency numbers
words, counts = zip(*top_20_words)

plt.figure(figsize=(12, 8))
plt.barh(words, counts, color='teal', edgecolor='black')
plt.gca().invert_yaxis() # Put the #1 word at the top
plt.title('Top 20 Most Frequent Words in Descriptions (Stop Words Removed)', fontsize=16)
plt.xlabel('Frequency', fontsize=12)
plt.show()
```

```python
# 1. Calculate Word Counts
# We simply split the string by spaces and count the list length

# splitting each discriptoin into list, by applying a lambda funciotn to each description
desc_lengths = all_titles_clean['description'].astype(str).apply(lambda x: len(clean_description(x).split()))

print(desc_lengths)

plt.figure(figsize=(10, 6))
plt.hist(desc_lengths, bins=20, color='purple', edgecolor='black', alpha=0.7)

mean_length = desc_lengths.mean()
plt.axvline(mean_length, color='red', linestyle='dashed', linewidth=2, label=f'Average: {int(mean_length)} words')

plt.title('Distribution of Description Word Counts', fontsize=16)
plt.xlabel('Number of Words', fontsize=12)
plt.ylabel('Count of Movies/Shows', fontsize=12)
plt.legend()

plt.show()
```

# Pre Processing


I realised that I need to make the description, listed_in, and title columns lower case, as I will be using them in the training part.

Also I will be making the user request in order to make a recomdnetion.

#### 2. Creating the soap column.
#### 3. Transform the soap into numeric.
#### 4. Reduce the dimentions to 2D for visualizaiton (plotting)

## 1. Cleaning the description and listed_in, and Title

```python
all_titles_clean = all_titles_clean.reset_index(drop=True)
all_titles_clean['clean_description'] = all_titles_clean['description'].apply(clean_description)
all_titles_clean['clean_listed_in'] = all_titles_clean['listed_in'].apply(clean_genres)
all_titles_clean['clean_title'] = all_titles_clean['title'].apply(clean_title)

```

## 2. Creating the soap column.

```python
# Making a soap column that will combine listed_in and description in order to convert them into numeric.
all_titles_clean['soup'] = (
    all_titles_clean['clean_title'] + " " + 
    # all_titles_clean['clean_title'] + " " +  # Repeated for weight!
    all_titles_clean['clean_listed_in'] + " " + 
    all_titles_clean['clean_description']
)
```

## 3. Transform the soap into numeric.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# I  used the build in stop_words='english', the result is not very good, thats why removed the stopwords manually
with_stop_words = all_titles_clean['listed_in'] + " " + all_titles_clean['description']
tfidf_stopwords = TfidfVectorizer(stop_words='english')
tfidf_stopwords_matrix = tfidf_stopwords.fit_transform(with_stop_words)

print("Matrix With stopwords: ", tfidf_stopwords_matrix.shape)

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(all_titles_clean['soup'])

print("Matrix Shape:", tfidf_matrix.shape)
```

## 4. Reduce the dimentions to 2D for visualizaiton (plotting)

```python
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD(n_components=2, random_state=42)
matrix_2d_with_stopwords = svd.fit_transform(tfidf_stopwords_matrix)
matrix_2d = svd.fit_transform(tfidf_matrix)


fig, axes = plt.subplots(1, 2, figsize=(16, 6))

axes[0].scatter(matrix_2d[:, 0], matrix_2d[:, 1], color='red',alpha=0.3, s=1)
axes[0].set_title('Streaming Universe With Cleaning', fontsize=14)
axes[0].set_xlabel('Component 1 (General Themes)', fontsize=12)
axes[0].set_ylabel('Component 2 (Niche Topics)', fontsize=12)



axes[1].scatter(matrix_2d_with_stopwords[:, 0], matrix_2d_with_stopwords[:, 1], color='blue', alpha=0.3, s=1)
axes[1].set_title('Streaming Universe Without Cleaning', fontsize=14)
axes[1].set_xlabel('Component 1 (General Themes)', fontsize=12)
axes[1].set_ylabel('Component 2 (Niche Topics)', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# This Scatter is for all TV Shows and Movies
# Defining the colors of the points corresponding to their platform
platform_colors = {
    'Netflix': 'red',
    'Amazon Prime': 'lightblue', # Check your exact spelling in the dataframe!
    'Disney Plus': 'blue',
    'Hulu': 'green'
}

plt.figure(figsize=(12, 10))


# Group the titles by platform.
groups = all_titles_clean.groupby('platform') # Now groups has 4 group each with name


# Iterating the groups and scatter them in the same figure.
for name, group in groups:

    # This ensure that we are slecting the right platform name for each platform from the matrix_2d
    mask = (all_titles_clean['platform'] == name) 
    # Apply the mask to the 2D matrix
    subset_matrix = matrix_2d[mask]
    
    # Get the specific color for this platform
    color = platform_colors.get(name)
    
    # PLOT
    # each loop iteratoin the name, color, and subset_matrix will be different, based on the platform.
    plt.scatter(subset_matrix[:, 0], subset_matrix[:, 1], 
                label=name,
                c=color,     
                alpha=0.4, 
                s=3)

plt.legend(title='Platform', markerscale=5)
plt.title('Streaming Content Universe by Platform', fontsize=16)
plt.xlabel('Component 1 (General Themes)', fontsize=12)
plt.ylabel('Component 2 (Niche Topics)', fontsize=12)
plt.grid(alpha=0.2)

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 9), sharey=True)

# There are 2 axis, first one is Movie, and second is TV Show
plot_settings = [('Movie', axes[0]), ('TV Show', axes[1])]

# iterating through Movie, TV Show and getting its type and axis.
for type, ax in plot_settings:
    # Filtering its type if Movie or TV Show
    type_mask = all_titles_clean['type'] == type
    # Applying the filter on the dataset
    subset_df = all_titles_clean[type_mask]
    # Applying the filter on t he matrix dataframe
    subset_matrix = matrix_2d[type_mask]

    # Also grouping them again based on platform
    groups = subset_df.groupby('platform')

    # getting the name and the group from the grouped by
    for name, group in groups:
        # filtering each platform based on coresponding platform
        platform_mask = subset_df['platform'] == name
        # applying the filter
        points = subset_matrix[platform_mask]
        # Putting the coresponding color.
        color = platform_colors.get(name)

        # Scatter the each platform in its axis with coresponding type
        ax.scatter(points[:, 0], points[:, 1], 
                   label=name, 
                   c=color, 
                   alpha=0.4, 
                   s=4)
    # after the first axis plotted, then setting title and lables.
    ax.set_title(f'The Universe of {type}s', fontsize=18)
    ax.set_xlabel('Component 1 (Themes)', fontsize=14)
    if type == 'Movie':
        ax.set_ylabel('Component 2 (Niches)', fontsize=14)
        ax.legend(title='Platform', markerscale=5, loc='upper left') # Legend only on first plot
    ax.grid(alpha=0.2)

plt.tight_layout()
plt.show()
```

# Training the model using K-NN

Usint the K-NN, to find the nearest neighbors of a given title, and also a given search term.

```python
from sklearn.neighbors import NearestNeighbors

# 1. Define the Model
# metric='cosine': Calculates 1 - cosine_similarity (distance)
# algorithm='brute': Standard for sparse data like text
knn = NearestNeighbors(metric='cosine', algorithm='brute')

# 2. Fit the Model (Load the High-Dim Matrix)
# Note: Use 'tfidf_matrix', NOT 'matrix_2d'
knn.fit(tfidf_matrix)

print("Model trained and ready for queries!")
```
